```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import tweepy 
import datetime

import matplotlib.dates as mdates
import matplotlib.cbook as cbook

# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Import Twitter API Keys
from config import consumer_key, consumer_secret, access_token, access_token_secret

# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
#target_users = ["BBC","CBS","CNN","FoxNews","nytimes"]

#BBC
user_BBC = "BBC"
user_name_BBC = []
compound_BBC = []
pos_BBC = []
neg_BBC = []
neu_BBC = []
time_BBC = []
tweetsago_BBC = []
counter = 1

public_tweets_BBC = api.user_timeline(user_BBC,count=100)

for tweet in public_tweets_BBC:
        #print(json.dumps(tweet, sort_keys=True, indent=4))
        user_name_BBC.append(tweet["user"]["name"])
        results = analyzer.polarity_scores(str(tweet["text"]))
        compound_BBC.append(results["compound"])
        pos_BBC.append(results["pos"])
        neg_BBC.append(results["neg"])
        neu_BBC.append(results["neu"])
        time_BBC.append(tweet["created_at"])
        tweetsago_BBC.append(counter)
        counter += 1
        
#CBS
user_CBS = "CBS"
user_name_CBS = []
compound_CBS = []
pos_CBS = []
neg_CBS = []
neu_CBS = []
time_CBS = []
tweetsago_CBS = []
counter = 1

public_tweets_CBS = api.user_timeline(user_CBS,count=100)

for tweet in public_tweets_CBS:
        #print(json.dumps(tweet, sort_keys=True, indent=4))
        user_name_CBS.append(tweet["user"]["name"])
        results = analyzer.polarity_scores(str(tweet["text"]))
        compound_CBS.append(results["compound"])
        pos_CBS.append(results["pos"])
        neg_CBS.append(results["neg"])
        neu_CBS.append(results["neu"])
        time_CBS.append(tweet["created_at"])
        tweetsago_CBS.append(counter)
        counter += 1

#CNN
user_CNN = "CNN"
user_name_CNN = []
compound_CNN = []
pos_CNN = []
neg_CNN = []
neu_CNN = []
time_CNN = []
tweetsago_CNN = []
counter = 1

public_tweets_CNN = api.user_timeline(user_CNN,count=100)

for tweet in public_tweets_CNN:
        #print(json.dumps(tweet, sort_keys=True, indent=4))
        user_name_CNN.append(tweet["user"]["name"])
        results = analyzer.polarity_scores(str(tweet["text"]))
        compound_CNN.append(results["compound"])
        pos_CNN.append(results["pos"])
        neg_CNN.append(results["neg"])
        neu_CNN.append(results["neu"])
        time_CNN.append(tweet["created_at"])
        tweetsago_CNN.append(counter)
        counter += 1

#FoxNews
user_FoxNews = "FoxNews"
user_name_FoxNews = []
compound_FoxNews = []
pos_FoxNews = []
neg_FoxNews = []
neu_FoxNews = []
time_FoxNews = []
tweetsago_FoxNews = []
counter = 1

public_tweets_FoxNews = api.user_timeline(user_FoxNews,count=100)

for tweet in public_tweets_FoxNews:
        #print(json.dumps(tweet, sort_keys=True, indent=4))
        user_name_FoxNews.append(tweet["user"]["name"])
        results = analyzer.polarity_scores(str(tweet["text"]))
        compound_FoxNews.append(results["compound"])
        pos_FoxNews.append(results["pos"])
        neg_FoxNews.append(results["neg"])
        neu_FoxNews.append(results["neu"])
        time_FoxNews.append(tweet["created_at"])
        tweetsago_FoxNews.append(counter)
        counter += 1
        
#nytimes
user_nytimes = "nytimes"
user_name_nytimes = []
compound_nytimes = []
pos_nytimes = []
neg_nytimes = []
neu_nytimes = []
time_nytimes = []
tweetsago_nytimes = []
counter = 1

public_tweets_nytimes = api.user_timeline(user_nytimes,count=100)

for tweet in public_tweets_nytimes:
        #print(json.dumps(tweet, sort_keys=True, indent=4))
        user_name_nytimes.append(tweet["user"]["name"])
        results = analyzer.polarity_scores(str(tweet["text"]))
        compound_nytimes.append(results["compound"])
        pos_nytimes.append(results["pos"])
        neg_nytimes.append(results["neg"])
        neu_nytimes.append(results["neu"])
        time_nytimes.append(tweet["created_at"])
        tweetsago_nytimes.append(counter)
        counter += 1
    
```

```python
#BBC
news_info_BBC = pd.DataFrame({"Username" : user_name_BBC, "Compound Score" : compound_BBC, "Positive Score" : pos_BBC,"Negative Score" : neg_BBC, "Neutral Score" : neu_BBC,"Time Posted" : time_BBC, "Tweets Ago" : tweetsago_BBC})
news_info_BBC.sort_values(by=["Time Posted"],ascending=False).head()

#CBS
news_info_CBS = pd.DataFrame({"Username" : user_name_CBS, "Compound Score" : compound_CBS, "Positive Score" : pos_CBS,"Negative Score" : neg_CBS, "Neutral Score" : neu_CBS, "Time Posted" : time_CBS, "Tweets Ago" : tweetsago_CBS})
news_info_CBS.sort_values(by=["Time Posted"],ascending=False).head()

#CNN
news_info_CNN = pd.DataFrame({"Username" : user_name_CNN, "Compound Score" : compound_CNN,"Positive Score" : pos_CNN,"Negative Score" : neg_CNN, "Neutral Score" : neu_CNN, "Time Posted" : time_CNN, "Tweets Ago" : tweetsago_CNN})
news_info_CNN.sort_values(by=["Time Posted"],ascending=False).head()

#FoxNews
news_info_FoxNews = pd.DataFrame({"Username" : user_name_FoxNews, "Compound Score" : compound_FoxNews, "Positive Score" : pos_FoxNews,"Negative Score" : neg_FoxNews, "Neutral Score" : neu_FoxNews, "Time Posted" : time_FoxNews, "Tweets Ago" : tweetsago_FoxNews})
news_info_FoxNews.sort_values(by=["Time Posted"],ascending=False).head()

#nytimes
news_info_nytimes = pd.DataFrame({"Username" : user_name_nytimes, "Compound Score" : compound_nytimes, "Positive Score" : pos_nytimes,"Negative Score" : neg_nytimes, "Neutral Score" : neu_nytimes, "Time Posted" : time_nytimes, "Tweets Ago" : tweetsago_nytimes})
news_info_nytimes.sort_values(by=["Time Posted"],ascending=False).head()
```

```python
#BBC
x_axis_BBC = news_info_BBC["Tweets Ago"]
y_axis_BBC = news_info_BBC["Compound Score"]
plt.scatter(x_axis_BBC,y_axis_BBC,marker='o',c="lightblue",edgecolors='k',label="BBC")

#CBS
x_axis_CBS = news_info_CBS["Tweets Ago"]
y_axis_CBS = news_info_CBS["Compound Score"]
plt.scatter(x_axis_CBS,y_axis_CBS,marker='o',c="green",edgecolors='k',label="CBS")

#CNN
x_axis_CNN = news_info_CNN["Tweets Ago"]
y_axis_CNN = news_info_CNN["Compound Score"]
plt.scatter(x_axis_CNN,y_axis_CNN,marker='o',c="red",edgecolors='k',label="CNN")

#FoxNews
x_axis_FoxNews = news_info_FoxNews["Tweets Ago"]
y_axis_FoxNews = news_info_FoxNews["Compound Score"]
plt.scatter(x_axis_FoxNews,y_axis_FoxNews,marker='o',c="blue",edgecolors='k',label="Fox News")

#nytimes
x_axis_nytimes = news_info_nytimes["Tweets Ago"]
y_axis_nytimes = news_info_nytimes["Compound Score"]
plt.scatter(x_axis_nytimes,y_axis_nytimes,marker='o',c="yellow",edgecolors='k',label="Fox News")

plt.ylim(-1,1)
timeNow = datetime.datetime.now().strftime("%m/%d/%Y")
plt.title(f"Sentiment Analysis as of {timeNow}")
plt.xlabel("Tweets Ago")
plt.ylabel("Compound Score")
#plt.legend()
plt.grid()
plt.xticks()


plt.savefig("Images/Compound_ScatterPlot.png")
plt.show()
```

```python
#BBC
avgBBC = np.mean(compound_BBC)

#CBS
avgCBS = np.mean(compound_CBS)

#CNN
avgCNN = np.mean(compound_CNN)

#FoxNews
avgFoxNews = np.mean(compound_FoxNews)

#nytimes
avgnytimes = np.mean(compound_nytimes)

overall_compound = [avgBBC,avgCBS,avgCNN,avgFoxNews,avgnytimes]
x_axis_overall = np.arange(len(overall_compound))
labels = ["BBC","CBS","CNN","FoxNews","nytimes"]
```

```python
bar_graph = plt.bar(x_axis_overall, overall_compound, color='b', align="center")
tick_locations = [value for value in x_axis_overall]
plt.xticks(tick_locations, labels)
plt.xlim(-0.75,len(x_axis_overall)-0.25)
plt.title(f"Overall Compound Score of News Tweets as of {timeNow}")
plt.xlabel("News Organizations")
plt.ylabel("Overall Compound Score")
bar_graph[0].set_color('r')
bar_graph[1].set_color('g')
bar_graph[2].set_color('y')
bar_graph[3].set_color('k')
for a,b in zip(x_axis_overall, overall_compound):
    plt.text(a, b, str(round(b,3)))
plt.savefig("Images/OverallCompound_BarGraph.png")
plt.show()
```
